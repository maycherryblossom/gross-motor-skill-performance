```python
import numpy as np
import pandas as pd

data = pd.read_csv("human_test.csv")
```

```python
data_2 = data[data["Labler 이름"]==""]
data_3 = data[data["Labler 이름"]==""]
data_4 = data[data["Labler 이름"]==""]
data_1 = data[data["Labler 이름"]==""]
```

```python
y_pred = data_1["최종 점수"]
y_true = data_1["최종 점수 정답"]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_true = encoder.fit_transform(y_true)
y_pred = encoder.fit_transform(y_pred)
```

```python
from sklearn.metrics import roc_auc_score, confusion_matrix

auc_score = roc_auc_score(y_true, y_pred)

tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print("AUC score:", auc_score)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
```
